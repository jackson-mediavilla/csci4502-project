```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import sklearn
from scipy import stats
import subprocess
import os
%matplotlib inline
```

```python
#writing a function that aggregates the data into csv files
paths = os.listdir('../AggregatePlayByPlays') #all the event files
start_year = 1950
end_year = 2017

#make csvs
for path in paths:

    t = path[-3:] #type of file. eva is event
    if(t!='EVA'): #if not game log just skip it
        continue
    
    y = int(path[:4]) #year in the path
    if(y>=start_year): #and past our start year
        command = str("bgame -y {0} {1} > {2}.txt".format(y, path, path[:7]))
        #print(command)
        try:
            subprocess.check_output(command,shell=True,stderr=subprocess.STDOUT)
        except subprocess.CalledProcessError as e:
            print("Finished {0}".format(path))#this is honestly just so i can see it's working
```
