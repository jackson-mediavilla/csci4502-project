```python
#writing a function that aggregates the data into csv files
#check_output("BEVENT.EXE -y 2007 -f 0-96 2007BOS.EVA", shell=True)
paths = 'paths.txt'#path file
pathsFile = open(paths, 'r')#open the file

#now going to try to make tons of CSV files
for line in pathsFile:
    #first format the line
    line = line[2:].rstrip()
    command = str("BEVENT.EXE -y {0} -f 0-96 {1} > {2}.csv".format(line[:4], line, line[:7]))
    try:
        subprocess.check_output(command,shell=True,stderr=subprocess.STDOUT)
    except subprocess.CalledProcessError as e:
        print("Finished {0}".format(line))#this is honestly just so i can see it's working
```
